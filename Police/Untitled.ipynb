```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Police_Data.csv")
```

```python
df.head()
```

```python
# Assuming df is your DataFrame
# Q. 1) Instruction ( For Data Cleaning ) - Remove the column that only contains missing values.
df.dropna(axis=1, how='all', inplace=True)
df.head()
```

```python
# Q. 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ? 
speeding_data = df[df['violation'] == 'Speeding']
speeding_counts_by_gender = speeding_data['driver_gender'].value_counts()
print(speeding_counts_by_gender)
```

```python
# Q. 3) Question ( Groupby ) - Does gender affect who gets searched during a stop ?
search_freq_by_gender = df.groupby('driver_gender')['search_conducted'].mean()
search_freq_by_gender.head()
```

```python
# Q. 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?
# Mapping stop_duration to numerical values
duration_mapping = {'0-15 Min': 7.5, '16-30 Min': 23, '30+ Min': 45}
df['stop_duration_numeric'] = df['stop_duration'].map(duration_mapping)

# Calculating mean stop_duration
mean_stop_duration = df['stop_duration_numeric'].mean()
print("Mean stop duration:", mean_stop_duration)
```

```python
# Q. 5) Question ( Groupby , Describe ) - Compare the age distributions for each violation
age_stats_by_violation = df.groupby('violation')['driver_age'].describe()
age_stats_by_violation.head()
```
